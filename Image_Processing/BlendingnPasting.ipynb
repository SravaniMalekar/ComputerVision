```python
#blend diff size images 
img = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/dog_backpack.png')
img1=  cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
imgg = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(imgg, cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.resize(img2,(600,600))
```

```python
img1.shape
```

```python
x_offset = 934 - 600
y_offset = 1401 - 600
```

```python
rows,columns,channel = img2.shape
```

```python
roi = img1[y_offset:1401,x_offset:934]
```

```python
plt.imshow(roi)
```

```python
img2gray = cv2.cvtColor(img2,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img2gray,cmap='gray')
```

```python
mask_inv = cv2.bitwise_not(img2gray)
```

```python
plt.imshow(mask_inv,cmap='gray')
```

```python
import numpy as np
```

```python
white_background = np.full(img2.shape,255,dtype=np.uint8)
```

```python
bk = cv2.bitwise_or(white_background,white_background, mask = mask_inv)
```

```python
plt.imshow(bk)
```

```python
fg = cv2.bitwise_or(img2,img2,mask=mask_inv)
```

```python
plt.imshow(fg)
```

```python
final_roi = cv2.bitwise_or(roi,fg)
```

```python
plt.imshow(final_roi)
```

```python
large_img = img1
small_img = final_roi
```

```python
large_img[y_offset:y_offset+small_img.shape[0],x_offset:x_offset+small_img.shape[1]] = small_img
```

```python
plt.imshow(large_img)
```
